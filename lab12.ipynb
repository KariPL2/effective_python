Krok 7: Przygotuj efektywny wykres na którym przedstawisz procetowy udział poszczególnych zawodów w zbiorze na charplocie
Wykorzystaj https://matplotlib.org/examples/pie_and_polar_charts/pie_demo_features.html

Przygotuj dwa wykresy: 
- jeden z wszystkimi zawodami
- drugi z 10 najcześtrzymi zawodami, pozostałem zawody trafiają do kategorii `Inne`

```python
import matplotlib.pyplot as plt




groupby_occ=ocupation.groupby(by="occupation")["occupation"].count()
groupby_occ.sort_values(axis=0)

labels=groupby_occ.keys()


fig,ax=plt.subplots()
ax.pie(groupby_occ,labels=labels,autopct='%1.1f%%')
plt.show()

```

Krok 8: Przygotuj efektywny wykres na którym przedstawisz 10 najczęstrzych zawodów oraz ile pracuje w nich kobiet a ilu meżczyzn
Wykorzystaj https://matplotlib.org/gallery/api/barchart.html#sphx-glr-gallery-api-barchart-py

```python

import numpy as np
import matplotlib.pyplot as plt

groupby_occ=ocupation["occupation"].value_counts()
# print(groupby_occ)
occupations = ocupation["occupation"].unique()
counts = {}
male_query = ocupation["gender"] == "M"
for occ in occupations:
    occ_query = ocupation["occupation"] == occ
    people = ocupation.loc[occ_query]
    males = people.loc[male_query]
    n_people =  len(people)
    n_males = len(males)
    n_females = n_people - n_males
    counts[occ] = (n_people, n_males, n_females)
    print(occ, n_males, n_females)

counts = dict(sorted(counts.items() , key = lambda pair:pair[1][0],reverse = True))

i=0

ten_males=[]
ten_females=[]
occ=[]
for k,v in counts.items():
    if i == 10: break
    occ.append(k)
    ten_males.append(counts[k][1])
    ten_females.append(counts[k][2])
    print ("zawod, mezcyzni {} {} ".format(k,counts[k][1])) 
    print ("zawod, kobiety {} {} ".format(k,counts[k][2]))
    i += 1

#print(counts)
ind = np.arange(len(ten_males))  # the x locations for the groups
width = 0.25  # the width of the bars
fig,ax=plt.subplots()
rects1=ax.bar(ind-width/2,ten_males,width,color="SkyBlue",label="Men")
rects2=ax.bar(ind+width/2,ten_females,width,color="IndianRed",label="Female")

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind,occ,fontsize=6)

ax.legend()

plt.show()

# groupby_occ=ocupation.groupby(["occupation","gender"])["occupation"]
# print(groupby_occ.apply())
# groupby_occ.sort_values()

# print(groupby_occ.groups)





```

Krok 9: Przygotuj efektywny wykres na którym przedstawisz średni wiek pracowników danego zawodu

```python
groupby_occ_age=ocupation.groupby("occupation")["age"].mean()
print(groupby_occ_age)
# groupby_occ.sort_values()

ind = np.arange(len(groupby_occ_age.index))  # the x locations for the groups
print(ind)

width = 0.1  # the width of the bars
fig,ax=plt.subplots()
rects1=ax.bar(ind-width/2,groupby_occ_age.values,width,color="SkyBlue",label="Count")


ax.set_xlabel('Occupation')
ax.set_ylabel('Average Age')  # zmiana osi y na x
ax.set_title('Average Age by Occupation')
ax.set_xticks(ind)
ax.set_xticklabels(groupby_occ_age.index,rotation=90)

ax.legend()

plt.show()
```

Krok 10: Przygotuj efektywny wykres na którym przedstawisz średni wiek kobiet i mężczyzn wykonujących dany zawód.

```python


groupby_occ_age_gender = ocupation.groupby(["occupation", "gender"])["age"].mean()

ind = np.arange(len(groupby_occ_age_gender.index.levels[0]))
#print(ind)
width = 0.15
fig, ax = plt.subplots()

mean_age_male = groupby_occ_age_gender.xs('M', level='gender')
mean_age_female = groupby_occ_age_gender.xs('F', level='gender')
mean_age_female = mean_age_female.reindex(mean_age_male.index, fill_value=0)
# print(len(mean_age_male.index))
# print((mean_age_female.index))
rects1 = ax.bar(ind - width/2, mean_age_male.values, width, color="SkyBlue", label="Male")
rects2 = ax.bar(ind + width/2, mean_age_female.values, width, color="Red", label="Female")


ax.set_xlabel('Occupation')
ax.set_ylabel('Average Age')  # zmiana osi y na x
ax.set_title('Average Age by Occupation')
ax.set_xticks(ind)
ax.set_xticklabels(groupby_occ_age.index,rotation=90)

ax.legend()

plt.show()
```
